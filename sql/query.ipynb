```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine("sqlite:///analise_vendas.db")
```

```python
customers = pd.read_excel("../dados/customers_big_refactory.xlsx")
```

```python
orders = pd.read_excel("../dados/orders_big_refactory.xlsx")
```

```python
customers.to_sql("customers", engine, if_exists="replace", index=False)
```

```python
orders.to_sql("orders", engine, if_exists="replace", index=False)
```

```python
query = ("SELECT o.order_id AS 'ID', c.name AS 'Nome', c.city 'Cidade',"
         " o.order_date AS 'Data', o.product AS 'Produto', o.quantity AS 'Quantidade', o.price AS 'PreÃ§o'"
         " FROM customers AS c, orders AS o WHERE o.customer_id = c.customer_id AND o.order_date > '2024-01-01'")
```

```python
final_tabela = pd.read_sql(query, con=engine)
```

```python
final_tabela.to_excel('../dados/tabela_final.xlsx', index=False)
```
